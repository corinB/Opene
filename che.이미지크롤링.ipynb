```python
import os
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# 크롤링할 웹 페이지 URL
base_url = "https://www.musinsa.com/mz/streetsnap?style_type=americancasual&_mon=&gender=&p={page_number}#listStart"

# 크롤링할 페이지 수
x = 10

# 저장할 디렉토리 생성
output_dir = 'amecaji_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Selenium WebDriver 설정
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Headless 모드로 실행
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 페이지를 로드하고 이미지를 다운로드하는 함수
def crawl_page(page_number):
    page_dir = os.path.join(output_dir, f'{page_number}_{page_number}')
    if not os.path.exists(page_dir):
        os.makedirs(page_dir)
    
    print(f"Crawling page {page_number}")
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    img_tags = soup.select('div.list-box.box img')

    for index, img in enumerate(img_tags):
        img_url = img.get('src')
        img_url = urljoin(base_url.format(page_number=page_number), img_url)

        try:
            img_data = requests.get(img_url).content
            img_name = f"{index}.webp"
            img_path = os.path.join(page_dir, img_name)
            with open(img_path, 'wb') as f:
                f.write(img_data)
            print(f"Downloaded and converted: {img_name}")
        except Exception as e:
            print(f"Failed to download {img_url}: {e}")

# 1부터 x까지 페이지를 반복하여 크롤링
for page_number in range(1, x + 1):
    try:
        url = base_url.format(page_number=page_number)
        print(f"Navigating to page {page_number}: {url}")
        driver.get(url)
        time.sleep(5)  # 페이지가 완전히 로드될 때까지 대기
        crawl_page(page_number)
    except Exception as e:
        print(f"Failed to navigate to page {page_number}: {e}")

driver.quit()
print("Image download and conversion to WebP complete.")
```
